```python
from google.colab.patches import cv2_imshow
import cv2
image =cv2.imread("/content/Image 1.jpg")
cv2_imshow(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edge=cv2.Canny(gray,100,200)
contours,_ = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Topilgan konturlarni rasmga chizish
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

# Rasmni ko'rsatish
if 'google.colab' in str(get_ipython()):
    cv2_imshow(edge)  # Google Colab uchun
else:
    cv2.imshow('Contours', image)  # Local kompyuter uchun

```

```python
blurred=cv2.GaussianBlur(gray,(5,5),0)
cv2_imshow(blurred)
```
